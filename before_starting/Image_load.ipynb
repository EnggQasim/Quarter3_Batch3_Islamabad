```python
arr = np.random.randint(0,255,size=(7,7))
arr
```

```python
filt = np.array([[0,-1,0],
               [-1,5,-1],
               [-1,0,-1]])
filt
```

```python
for r in range(arr.shape[0]-2):
    for c in range(arr.shape[1]-2):
        print(r,c)
        print(arr[r:r+3,c:c+3])
   
```

```python
for r in range(arr.shape[0]-2):
    for c in range(arr.shape[1]-2):
        print(r,c)
        print(arr[r:r+3,c:c+3]*filt)
   
```

```python
for r in range(arr.shape[0]-2):
    for c in range(arr.shape[1]-2):
        #print(r,c)
        print((arr[r:r+3,c:c+3]*filt).sum())
```

```python
np.array((arr.shape)) - 2
```

```python
arr1 = np.zeros(np.array((arr.shape)) - 2)
for r in range(arr.shape[0]-2):
    for c in range(arr.shape[1]-2):
        #print(r,c)
        arr1[r][c]=(arr[r:r+3,c:c+3]*filt).sum()
arr1        
```

```python
arr2 = []#np.zeros((arr1.shape[0]//2, arr1.shape[1]//2))

for r in range(0,arr1.shape[0]-2,2):
    for c in range(0,arr1.shape[1]-2,2):
        arr2.append(arr1[r:r+2,c:c+2].max())
        
        
arr2 = np.array(arr2).reshape((arr1.shape[0]//2, arr1.shape[1]//2))
arr2
```
